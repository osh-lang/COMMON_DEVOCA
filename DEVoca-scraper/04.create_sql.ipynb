```python
import json

from pprint import pprint
from tqdm import tqdm
```

```python
with open('03.word_translated_link_categorised.json', 'r') as f:
    data = json.load(f)
    pprint(data[:3])
```

```python
categories = set()

for word in tqdm(data):
    categories.update(word['wordCategory'])

categories = sorted(categories)
pprint(categories)
print(len(categories))
```

```python
alphanumerics = "abcdefghijklmnopqrstuvwxyz0123456789"

with open('04.create_sql.sql', 'w') as f:
    f.write("USE DEVOCA;\n\n")

    for cat in tqdm(categories):
        f.write(f"INSERT INTO `categories`(category_name) VALUES ('{cat}');\n")
        f.write(f"SET @cat_{cat} = (SELECT LAST_INSERT_ID());\n")

    f.write("\n\n\n")

    for word in tqdm(data):
        f.write(f"INSERT INTO `words`(word_name_kr, word_name_en, word_def, word_link, word_summ) "
                f"VALUES (\'{word['word_name_kr']}\', \'{word['word_name_en']}\', "
                f"\'{word['word_def'].replace("\n", "<br>").replace("'", "&quot;")}\', "
                f"\'{word['word_link']}\', \'{word['word_summ'] if 'word_summ' in word else ''}\');\n")
        word_alpha_numeric = ''.join(filter(lambda x: x in alphanumerics, word['word_name_en']))
        f.write(f"SET @word_{word_alpha_numeric} = (SELECT LAST_INSERT_ID());\n")
        
        for cat in word['wordCategory']:
            f.write(f"INSERT INTO `words_categories`(wc_word_id, wc_category_id) "
                    f"VALUES (@word_{word_alpha_numeric}, @cat_{cat});\n")
```
